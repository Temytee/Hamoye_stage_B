```python
gun_types = {}

def gun_count(gun_type):
    if str(gun_type) != 'nan':
        guns = gun_type.split("||")
        if len(guns) == 1:
            guns = guns[0].split("|")
        for gun in guns:
            try:
                if gun.split('::')[1] in gun_types.keys():
                    gun_types[gun.split('::')[1]] += 1
                else:
                    gun_types[gun.split('::')[1]] = 0
            except:
                if gun.split(':')[1] in gun_types.keys():
                    gun_types[gun.split(':')[1]] += 1
                else:
                    gun_types[gun.split(':')[1]] = 0

violence_df['gun_type'].apply(gun_count);
```
